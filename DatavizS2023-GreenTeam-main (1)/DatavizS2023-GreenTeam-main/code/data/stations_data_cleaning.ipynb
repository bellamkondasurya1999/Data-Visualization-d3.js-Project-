```python
import os

directory = 'sea_level_data'
df = pd.DataFrame()
 
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    df1 = pd.read_table(directory + '/' + filename, header = None, sep=';')
    filename = directory + '/' + filename
    df1.columns = ['A', 'B', 'C', 'D']
    df1['Year'] = df1['A'].astype(int)
    df1 = df1[(df1['Year'] >= 2012)]
    df1 = df1[(df1['Year'] <= 2017)]
    df1['Month'] = (((df1['A'] % 1) * 12) + 1).astype(int)
    df1['Date'] = df1['Month'].astype(str) + "/1/" + df1['Year'].astype(str)
    df1['Station'] = filename[filename.index('/') + 1:filename.index('.')]
    df1 = df1.drop(['A', 'C', 'D', 'Year', 'Month'], axis=1)
    df1 = df1[['Station', 'Date', 'B']]
    df1.columns = ['Station', 'Date', 'Sea Level']
    df = pd.concat([df, df1])
    
stations = pd.read_csv('Stations.csv')
stations.columns = ['Station', 'City', 'Latitude', 'Longitude']
df['Station'] = df['Station'].astype(int)
df = pd.merge(df, stations, on=["Station"])
df = df[['Station', 'City', 'Longitude', 'Latitude', 'Date', 'Sea Level']]
    
```

```python
df.to_csv(index=False)
compression_opts = dict(method='zip', archive_name='out1.csv')  
df.to_csv('out1.zip', index=False, compression=compression_opts) 
```
